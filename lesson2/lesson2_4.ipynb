## 練習題：計算體重指數 BMI

BMI（Body Mass Index，身體質量指數）是評估體重是否健康的常用指標。

**計算公式：**  
BMI = 體重(公斤) ÷ 身高(公尺)²

**題目要求：**
1. 建立變數 `weight` 儲存體重 72 公斤
2. 建立變數 `height` 儲存身高 175 公分（注意：公式需要公尺，請先除以 100）
3. 使用 `**` 運算子計算身高的平方
4. 計算 BMI 值並儲存在變數 `bmi` 中
5. 使用 `round()` 函式將 BMI 四捨五入到小數點後 2 位
6. 印出結果

**參考輸出：**  
`23.35`

**提示：**
- 身高需轉換成公尺：`height = 175 / 100`
- 平方使用 `**` 運算子：`height ** 2`
- 四捨五入：`round(bmi, 2)`
