```python
#巢狀判斷
age = 50
if age < 18:
    print("未滿18")
    print("青少年")
else:
    if age < 40:
        print("青年")
    else:
        print("壯年")
```

```python
#多項選擇
age = 30
if age < 18:
    print("未滿18")
    print("青少年")
elif age < 40:
    print("青年")
else:
    print("壯年")
```
